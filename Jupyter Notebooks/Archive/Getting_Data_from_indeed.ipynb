# Reading a URL using `urllib`

More info: https://docs.python.org/3/howto/urllib2.html

```python
import urllib.request
url = "https://www.indeed.com.sg/jobs?q=data%20scientist&l=Central%20Business%20District"
request = urllib.request.Request(url)
response = urllib.request.urlopen(request)

html = response.read().decode('unicode_escape')
```

```python
import bs4
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html, "html.parser")
```

### extract job title
- Each job posting is nested under $\lt$div$\gt$ tags, wtih an attribute "class"="jobsearch-SerpJobCard row result"
- Then, job titles are listed under $\lt$a$\gt$ tags, with atrribute "title"

Therefore, to extract job title:
1. detect all $\lt$div$\gt$ tags with class containing "row"
2. then, find $\lt$a$\gt$ tags with attribute "data-tn-element":"jobTitle"
3. for all of these $\lt$a$\gt$ tags, extract the value of attribute "title"


```python
def extract_job_title_from_result(soup): 
    jobs = []
    for div in soup.find_all(name="div", attrs={"class":"row"}):
        for a in div.find_all(name="a", attrs={"data-tn-element":"jobTitle"}):
            jobs.append(a["title"])
    return(jobs)
jobnames = extract_job_title_from_result(soup)
```

### extract company name
- Each job posting is nested under $\lt$div$\gt$ tags, wtih an attribute "class"="jobsearch-SerpJobCard row result"
- Then, company are listed under $\lt$span$\gt$ tags, with atrribute "class=company"

Therefore, to extract job title:
1. detect all $\lt$div$\gt$ tags with class containing "row"
2. then, find $\lt$span$\gt$ tags with attribute "class":"company"
3. for all of these $\lt$span$\gt$ tags, extract the value

```python
def extract_company_from_result(soup): 
    companies = []
    for div in soup.find_all(name="div", attrs={"class":"row"}):
        company = div.find_all(name="span", attrs={"class":"company"})       
        for b in company:
            companies.append(b.text.strip())
    return(companies)
 
companynames = extract_company_from_result(soup)
```

```python
import pandas as pd
job_info = pd.DataFrame({'jobtitle':jobnames, 'companyname': companynames})
```
